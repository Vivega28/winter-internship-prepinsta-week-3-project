**Q3) Outliers: Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?**

```python
import numpy as np

# Define the z-score threshold
z_score_threshold = 3

# Calculate z-scores for Age, Salary, and Rating columns
z_scores = np.abs((data[['Age', 'Salary', 'Rating']] - data[['Age', 'Salary', 'Rating']].mean()) / data[['Age', 'Salary', 'Rating']].std())

# Filter out rows where all z-scores are less than or equal to the threshold
data_no_outliers = data[(z_scores <= z_score_threshold).all(axis=1)]

# Print the dataset after removing outliers
print("Dataset after removing outliers:")
print(data_no_outliers)

```

**Q4). Salary Formatting: Examine the format of the Salary column. Does it require any formatting or standardization for consistent analysis?**

```python
# Split the Salary column into minimum and maximum values
data[['Min Salary', 'Max Salary']] = data['Salary'].str.split('-', expand=True)

# Remove any '$' and 'k' symbols and convert to integers
data['Min Salary'] = data['Min Salary'].str.replace('$', '').str.replace('k', '000').astype(int)
data['Max Salary'] = data['Max Salary'].str.replace('$', '').str.replace('k', '000').astype(int)

# Display the updated DataFrame with minimum and maximum salary values
print("Updated DataFrame with minimum and maximum salary values (as integers):")
print(data[['Min Salary', 'Max Salary']])

```

14 Handling Categorical Data:
Question: For categorical indicators, consider encoding or transforming them into a format suitable for analysis.


```python
import pandas as pd

# Example DataFrame with categorical variable 'Category' and target variable 'Target'
data = pd.DataFrame({'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
                     'Target': [1, 0, 1, 1, 0, 1]})

# Calculate mean target value for each category
mean_target = data.groupby('Category')['Target'].mean()

# Map mean target values to corresponding categories
data['Category_Target_Encoded'] = data['Category'].map(mean_target)

print("Target Encoded DataFrame:")
print(data)
```

 15. Consistent Rating Scale:
Question: Ensure a consistent rating scale in the Rating column. Should it be normalized or adjusted for uniform analysis?

```python
# Assuming 'data' is your DataFrame and 'Rating' is the column containing ratings

# Import necessary libraries
import pandas as pd

# Sample DataFrame
data = pd.DataFrame({'Rating': [3.5, 4.2, 2.8, 5.0, 4.6, 2.0]})

# Calculate the maximum and minimum ratings
max_rating = data['Rating'].max()
min_rating = data['Rating'].min()

# Normalize ratings to a scale of 0 to 10
data['Normalized_Rating'] = 10 * (data['Rating'] - min_rating) / (max_rating - min_rating)

# Print the updated DataFrame
print("Updated DataFrame with normalized ratings:")
print(data)

```
